```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('world_country_stats.csv')
data.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = data['median_age'],  multiple='stack', bins=22)
plt.title('Histogram of Median Age Among the countries')
plt.xlabel('Median Age')
plt.ylabel('Number of Countries')
plt.yticks(range(0, 21, 2))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(data['region'],data['median_age'])
plt.title('Average age based on Region')
plt.xlabel('Region')
plt.ylabel('Median age')
```
